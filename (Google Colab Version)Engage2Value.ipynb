## DUMMY REGRESSOR

```python
X_train_d = train_data.drop('purchaseValue', axis=1)
y_train_d = train_data['purchaseValue']
X_test_d = test_data.drop('purchaseValue', axis=1, errors='ignore')
```

```python
model_1 = DummyRegressor()
```

```python
model_1.fit(X_train_d, y_train_d)
test_pred_d = model_1.predict(X_test_d)
```

```python
test_data['predicted_purchaseValue'] = test_pred_d
output_df = pd.DataFrame({
    'id': test_data.index,  # Index of the test data
    'purchaseValue': test_pred_d  # Predicted purchaseValue
})

```
